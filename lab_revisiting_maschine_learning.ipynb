```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
# Load data
data = pd.read_csv('learningSet.csv')
data.head()
```

```python
# Select categoricals
cat = data.select_dtypes(object)
cat.head()
```

```python
# Check for null values in all the columns
cat.isna().sum()
```

```python
# Null values don't look right
cat = cat.apply(lambda x: x.replace(" ", np.NaN))
cat.isna().sum()
```

```python
# Create a new empty list called drop_list and append `OSOURCE` and `ZIP CODE` to the list
drop_list = ['OSOURCE', 'ZIP']
```

```python
# Identify columns that over 85% missing values
nulls_percentage = cat.isna().sum()*100/len(cat)
nulls_85 = nulls_percentage[nulls_percentage > 85].index.tolist()
nulls_85
```

```python
# Remove those columns from the dataframe
drop_list.extend(nulls_85)
print(drop_list)
```

```python
cat = cat.drop(drop_list ,axis=1)
cat.head()
```

```python
# Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest
print(cat['GENDER'].value_counts())
cat['GENDER'] = cat['GENDER'].fillna('F')
```

```python
cat.loc[~cat['GENDER'].isin(['M', 'F']), 'GENDER'] = 'other'
print(cat['GENDER'].value_counts())
```
